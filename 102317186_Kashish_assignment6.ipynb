```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = Mx^2$', linestyle='-', color='blue')
plt.plot(x, y2, label=r'$y = Msin(x)$', linestyle='--', color='red')

plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Plot of Functions y = Mx^2 and y = Msin(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import seaborn as sns


data = {'Subjects': ['Math', 'Science', 'English', 'History', 'Computer Science'],
        'Scores': [85, 90, 78, 88, 95]}

df = pd.DataFrame(data)


plt.figure(figsize=(8, 5))
sns.barplot(x='Subjects', y='Scores', data=df, palette='muted')


for i, score in enumerate(df['Scores']):
    plt.text(i, score + 1, str(score), ha='center', fontsize=12)

plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Student Scores in Different Subjects")
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

data = np.random.rand(50)


fig, ax = plt.subplots(2, 2, figsize=(10, 6))


ax[0, 0].plot(np.cumsum(data), linestyle='-', marker='o', color='b')
ax[0, 0].set_title("Cumulative Sum")
ax[0, 0].grid()


noise = np.random.normal(0, 0.1, 50)
ax[0, 1].scatter(range(50), data + noise, color='r')
ax[0, 1].set_title("Scatter Plot with Random Noise")
ax[0, 1].grid()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv"


df = pd.read_csv(url)


print(df.head())


plt.figure(figsize=(8, 5))
sns.lineplot(x=df.index, y=df["total_profit"], marker="o", color="g")
plt.xlabel("Months")
plt.ylabel("Total Profit")
plt.title("Total Profit per Month")
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
for col in df.columns[1:-1]:
    sns.lineplot(x=df.index, y=df[col], label=col)

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Product Sales Data")
plt.legend()
plt.grid()
plt.show()


df.plot(kind="bar", figsize=(12, 6))
plt.title("Company Sales Data")
plt.xlabel("Months")
plt.ylabel("Values")
plt.grid()
plt.show()
```
