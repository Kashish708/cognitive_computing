```python
import numpy as np
import matplotlib.pyplot as plt

# Q1:
gfg = np.array([[4, 1, 9], [12, 3, 1], [4, 5, 6]])
print("Sum of all elements:", np.sum(gfg))
print("Sum of all elements row-wise:", np.sum(gfg, axis=1))
print("Sum of all elements column-wise:", np.sum(gfg, axis=0))
```

```python
# Q2:
a = np.array([10, 52, 62, 16, 16, 54, 453])
print("Indices of sorted array:", np.argsort(a))
print("Sorted array:", np.sort(a))
print("5 largest elements:", np.sort(a)[-5:])

b = np.array([[1.0, 1.2, 2.2], [2.0, 3.0, 2.0]])
print("Integer elements only:", b[b == b.astype(int)])
print("Float elements only:", b[b != b.astype(int)])
```

```python
# Q3:

X = sum(ord(c) for c in "K")
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales data:", sales)


tax_rate = ((X % 5) + 5) / 100
final_sales = sales * (1 + tax_rate)
print("Sales after tax:", final_sales)


discounted_sales = np.where(sales >= X+100, sales * 0.9, np.where(sales < X +100, sales * 0.95, sales))
print("Discounted sales:", discounted_sales)


sales_matrix = np.tile(sales, (3, 1))
sales_matrix = sales_matrix * (1.02 ** np.arange(3).reshape(-1, 1))
print("Expanded sales data:", sales_matrix)
```

```python
# Q4:
x = np.linspace(-10, 10, 100)
functions = {
    "Y = x^2": x**2,
    "Y = sin(x)": np.sin(x),
    "Y = e^x": np.exp(x),
    "Y = log(x+1)": np.log(x + 1)
}

for label, y in functions.items():
    plt.figure()
    plt.plot(x, y, label=label)
    plt.title(label)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid()
    plt.show()
```
